```python
from tkinter import *
from tkinter import ttk
from tkinter.filedialog import askopenfilename
from PIL import ImageTk, Image


bgcolor = "#0078D7"
lbcolor = "#0078D7"
btcolor = "#bcd4e6"


def createStyleWin():
    style_win = Tk() 
    style_win.title("Welcome")
    style_win.minsize(600, 400)
    style_win.configure(background=bgcolor)
    
    def OpenFile():
        name = askopenfilename(
                           filetypes =(("JPG", "*.jpg"),("All Files","*.*")),
                           title = "Choose a file."
                           )
        label.configure(text = name)
        print (name)
        #Using try in case user types in unknown file or closes without choosing a file.
        try:
            with open(name,'r') as UseFile:
                print(UseFile.read())
        except:
            print("No file exists")
    
    
    button_logout = Button(style_win, text = "Logout")
    button_logout.grid(column=3, row=0)
    button_logout.configure(background=btcolor)

    label_title = Label(style_win, text = "Style an Image")
    label_title.grid(column=0, row=0)
    label_title.configure(background=lbcolor)

    label_instr = Label(style_win, text = "Upload an image, then choose a style.")
    label_instr.grid(column=0, row=1)
    label_instr.configure(background=lbcolor)

    button_file = Button(style_win, text = "Upload an Image", command = OpenFile)
    button_file.grid(column=0, row=2)
    button_file.configure(background=btcolor)

    styles = {'Van Gogh', 'Picasso', 'Monet'}
    style_var = StringVar(style_win)
    style_var.set('Picasso') # set the default option
    popupMenu = OptionMenu(style_win, style_var, *styles)
    popupMenu.grid(column=1, row=2)
    popupMenu.configure(background=btcolor)
    
    #display images
    #tempimg = ImageTk.PhotoImage(Image.open("D:\images\image-placeholder.jpg"))
    
    #add image placeholder
    tempimg=Image.open('D:\images\image-placeholder.jpg')
    tempimg=tempimg.resize((100,80),Image.ANTIALIAS)
    tempimg=ImageTk.PhotoImage(tempimg)
    l1 = Label(image=tempimg)
    l1.grid(column=0,row=3)
    
    stylesloc = ["D:\images\vangogh.jpg", "D:\images\picasso.jpg", "D:\images\monet.jpg"]
    
    choosestyle = stylesloc[1]
    styleimg=Image.open(choosestyle)
    styleimg=styleimg.resize((100,80),Image.ANTIALIAS)
    styleimg=ImageTk.PhotoImage(styleimg)
    l2 = Label(image=styleimg)
    l2.grid(column=1,row=3)

    def change_dropdown(*args):
        print( style_var.get() )

    

    label = Label(style_win, text = "File Path Goes Here")
    label.grid(column=0, row=5)
    label.configure(background=lbcolor)

    style_win.mainloop()

    
createStyleWin()
```
